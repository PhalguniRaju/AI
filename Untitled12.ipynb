```python
keywords = {"if", "else", "while", "for", "return", "class", "def", "import", "True", "False"}

operators = {"+", "-", "*", "/", "%", "=", "==", "!=", ">", "<", ">=", "<="}

special_symbols = {"(", ")", "{", "}", "[", "]", ",", ";", ":"}

token = input("Enter a token: ")

if token in keywords:
    print("Keyword")
elif token in operators:
    print("Operator")
elif token in special_symbols:
    print("Special Symbol")
elif token.isidentifier():
    print("Identifier")
elif token.isnumeric():
    print("Constant (Number)")
else:
    print("Unknown Token")
```
